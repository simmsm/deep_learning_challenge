```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
```

```python
# Drop the non-beneficial ID columns, 'EIN'.
application_df= application_df.drop(columns=['EIN'], axis=1)
application_df.head()
```

```python
value_counts_name= application_df['NAME'].value_counts()
print(len(value_counts_name))
print(len(value_counts_name[value_counts_name>1]))
```

```python
# Choose a cutoff value and create a list of NAme types to be replaced
# use the variable name `NAME_to_replace`

num_values_to_combine= (19568-793)
application_types_to_replace = value_counts_name.nsmallest(num_values_to_combine).index


# Replace in dataframe
for app in application_types_to_replace:
    application_df['NAME'] = application_df['NAME'].replace(app,"Other")

# Check to make sure binning was successful
application_df['NAME'].value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
value_counts=application_df['APPLICATION_TYPE'].value_counts()

num_values_to_combine= 10
application_types_to_replace = value_counts.nsmallest(num_values_to_combine).index


# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
classification_value_counts= application_df['CLASSIFICATION'].value_counts()

# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
num_values_to_combine= 63
classifications_to_replace = classification_value_counts.nsmallest(num_values_to_combine).index

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
application_df1= pd.get_dummies(application_df, columns = ['NAME','APPLICATION_TYPE','AFFILIATION','CLASSIFICATION','USE_CASE','ORGANIZATION','INCOME_AMT','SPECIAL_CONSIDERATIONS'])
application_df1.head()
```

```python
# Split our preprocessed data into our features and target arrays
X = application_df1.copy()
X.drop("IS_SUCCESSFUL", axis=1, inplace=True)

y = application_df1["IS_SUCCESSFUL"].values.reshape(-1, 1)

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.

nn_model = tf.keras.models.Sequential()

# First hidden layer
nn_model.add(tf.keras.layers.Dense(units=1000, activation="LeakyReLU", input_dim= X.shape[1]))
# Second hidden layer
nn_model.add(tf.keras.layers.Dense(units=500, activation="sigmoid"))
# Output layer
nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))
# Check the structure of the model
nn_model.summary()
```

```python
# Compile the model
nn_model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
 #Train the model
callback= tf.keras.callbacks.ModelCheckpoint(filepath='modelcheckpoint',verbose=0, save_weights_only=True, save_freq=5)
fit_model = nn_model.fit(X_train_scaled, y_train, epochs=12, callbacks=[callback])
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
